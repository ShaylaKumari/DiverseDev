# EXTRAÇÃO DE DADOS DA API

```python
import requests 
import pandas as pd

url = "https://servicodados.ibge.gov.br/api/v1/localidades/municipios"

data = requests.get(url).json()
```

# CRIAÇÃO DE DATA FRAME

```python
df = pd.json_normalize(data, sep="_")

df = df.rename(columns={
    "nome": "municipio",
    "microrregiao_mesorregiao_UF_nome": "uf",
    "regiao-imediata_regiao-intermediaria_UF_sigla": "uf_sigla",
    "regiao-imediata_regiao-intermediaria_UF_regiao_nome": "regiao",
    "regiao-imediata_regiao-intermediaria_UF_regiao_sigla": "regiao_sigla"
})

df.head()
```

# FILTROS

```python
uf = "BA"
df_uf = df[df["uf_sigla"] == uf]
df_uf.head()
```

```python
regiao = "Nordeste"
df_reg = df[df["regiao"] == regiao]
df_reg.head()
```

```python
qntd_municipios_por_uf = df.groupby(["uf_sigla"])["id"].count().reset_index(name="n_municipios")
qntd_municipios_por_uf = qntd_municipios_por_uf.sort_values("n_municipios", ascending=False)
qntd_municipios_por_uf
```

```python
qntd_municipios_por_regiao = df.groupby("regiao")["id"].count().reset_index(name="n_municipios")
qntd_municipios_por_regiao = qntd_municipios_por_regiao.sort_values("n_municipios", ascending=False)
qntd_municipios_por_regiao
```

```python
duplicados = df.groupby("municipio").filter(lambda x: x["uf_sigla"].nunique() > 1)

homonimos = (
    duplicados.groupby("municipio")["uf_sigla"]
    .agg([
        ("uf", lambda x: sorted(x.unique())),
        ("n_ufs", lambda x: x.nunique())
    ])
    .reset_index()
    .sort_values("n_ufs", ascending=False)
)

homonimos
```

# DOWNLOAD

```python
df.to_csv("municipios_BR.csv", index=False, encoding="utf-8")
df_uf.to_csv("municipios_UF.csv", index=False, encoding="utf-8")
df_reg.to_csv("municipios_regiao_informada.csv", index=False, encoding="utf-8")
qntd_municipios_por_uf.to_csv("qntd_municipios_por_uf.csv", index=False, encoding="utf-8")
qntd_municipios_por_regiao.to_csv("qntd_municipios_por_regiao.csv", index=False, encoding="utf-8")
duplicados.to_csv("duplicados.csv", index=False, encoding="utf-8")
```

# ANÁLISES

```python
import matplotlib.pyplot as plt
```

```python
qntd_municipios_por_regiao = df.groupby("regiao")["id"].count().reset_index(name="qtde")
qntd_municipios_por_regiao = qntd_municipios_por_regiao.sort_values("qtde", ascending=True)

plt.figure(figsize=(7,5))
plt.barh(qntd_municipios_por_regiao["regiao"], qntd_municipios_por_regiao["qtde"])
plt.title("Quantidade de municípios por região")
plt.xlabel("Número de municípios")
plt.ylabel("Região")
plt.show()
```

```python
qntd_municipios_por_uf = df.groupby("uf_sigla")["id"].count().reset_index(name="qtde")
todas_uf = qntd_municipios_por_uf.sort_values("qtde", ascending=False)

plt.figure(figsize=(15,5))
plt.bar(todas_uf["uf_sigla"], todas_uf["qtde"])
plt.title("Distribuição de municípios por UF")
plt.xlabel("UF")
plt.ylabel("Número de municípios")
plt.show()
```

```python
top15 = homonimos.sort_values("n_ufs", ascending=True).tail(15)

plt.figure(figsize=(7, 5))
plt.barh(top15["municipio"], top15["n_ufs"])
plt.title("top 15 nomes de municípios repetidos nos estados")
plt.xlabel("Número de estados onde o nome aparece")
plt.ylabel("Nome do município")
plt.show()
```
