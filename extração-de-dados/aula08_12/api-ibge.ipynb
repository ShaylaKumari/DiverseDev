# EXTRAÇÃO DE DADOS DA API

```python
import requests 
import pandas as pd

url = "https://servicodados.ibge.gov.br/api/v1/localidades/municipios"

data = requests.get(url).json()
```

# CRIAÇÃO DE DATA FRAME

```python
df = pd.json_normalize(data, sep="_")

df = df.rename(columns={
    "nome": "municipio",
    "microrregiao_mesorregiao_UF_nome": "uf",
    "regiao-imediata_regiao-intermediaria_UF_sigla": "uf_sigla",
    "regiao-imediata_regiao-intermediaria_UF_regiao_nome": "regiao",
    "regiao-imediata_regiao-intermediaria_UF_regiao_sigla": "regiao_sigla"
})

df.head()
```

# FILTROS

```python
uf = "BA"
df_ba = df[df["uf_sigla"] == uf]
df_ba.head()
```

```python
regiao = "Nordeste"
df_reg = df[df["regiao"] == regiao]
df_reg.head()
```

```python
distribuiçao_mun = df.groupby(["uf_sigla"])["id"].count().reset_index(name="n_municipios")
distribuiçao_mun = distribuiçao_mun.sort_values("n_municipios", ascending=False)
distribuiçao_mun
```

```python
qntd_regiao = df.groupby("regiao")["id"].count().reset_index(name="n_municipios")
qntd_regiao = qntd_regiao.sort_values("n_municipios", ascending=False)
qntd_regiao
```

```python
duplicados = df.groupby("municipio").filter(lambda x: x["uf_sigla"].nunique() > 1)

homonimos = (
    duplicados.groupby("municipio")["uf_sigla"]
    .agg([
        ("uf", lambda x: sorted(x.unique())),
        ("n_ufs", lambda x: x.nunique())
    ])
    .reset_index()
    .sort_values("n_ufs", ascending=False)
)

homonimos
```

# DOWNLOAD

```python
df.to_csv("municipios_BR.csv", index=False, encoding="utf-8")
df_ba.to_csv("municipios_BA.csv", index=False, encoding="utf-8")
df_reg.to_csv("municipios_regiao_especifica.csv", index=False, encoding="utf-8")
distribuiçao_mun.to_csv("municipios_por_uf.csv", index=False, encoding="utf-8")
qntd_regiao.to_csv("qntd_municipios_por_regiao.csv", index=False, encoding="utf-8")
duplicados.to_csv("duplicados.csv", index=False, encoding="utf-8")
```

# ANÁLISES

```python
import matplotlib.pyplot as plt
```

```python
qntd_regiao = df.groupby("regiao")["id"].count().reset_index(name="qtde")
qntd_regiao = qntd_regiao.sort_values("qtde", ascending=True)

plt.figure(figsize=(7,5))
plt.barh(qntd_regiao["regiao"], qntd_regiao["qtde"])
plt.title("Quantidade de municípios por região")
plt.xlabel("Número de municípios")
plt.ylabel("Região")
plt.show()
```

```python
distribuiçao_mun = df.groupby("uf_sigla")["id"].count().reset_index(name="qtde")
todas_uf = distribuiçao_mun.sort_values("qtde", ascending=False)

plt.figure(figsize=(15,5))
plt.bar(todas_uf["uf_sigla"], todas_uf["qtde"])
plt.title("Distribuição de municípios por UF")
plt.xlabel("UF")
plt.ylabel("Número de municípios")
plt.show()
```

```python
top = homonimos.sort_values("n_ufs", ascending=True).tail(15)

plt.figure(figsize=(7, 5))
plt.barh(top["municipio"], top["n_ufs"])
plt.title("Top 15 nomes de municípios repetidos nos estados")
plt.xlabel("Número de estados onde o nome aparece")
plt.ylabel("Nome do município")
plt.show()
```
